```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("proj_data.xlsx")
weights = np.array([0.1] * 10) 


combined_return = df.dot(weights)


combined_index = (1 + combined_return).cumprod()


plt.figure(figsize=(10, 5))
plt.plot(combined_index, label='Combined Index', color='blue')
plt.title("Combined Index of 10 Assets (Equal Weight)")
plt.xlabel("Time (Days)")
plt.ylabel("Index Value")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
